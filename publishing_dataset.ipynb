# Publishing dataset in Hugging Face
## Instructions 
1. Generate a token in Hugging Face
1. create a *.env* file and create the following environment variable:
    - HF_HUB_TOKEN
    

```python
from datasets import load_dataset
```

```python
ds = load_dataset("./images/", split="train")
```

```python
from os import getenv
from dotenv import load_dotenv

load_dotenv()
ds.push_to_hub("oswaldo-guayasamin-blip-captions-v2",token=getenv("HF_HUB_TOKEN"))
```
